```python
#importing the necessary library 
import pandas as pd
```

```python
#Read the dataset file
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#Convert the categorical data to NUmerical Data

dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#List the number of columns 
dataset.columns
```

```python
#Input and output split 

independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
independent
```

```python
dependent
```

```python
#Training and Test data Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Data pre-processing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Model creation 
#Importing GridSearchCV for tuning the hyperparameters 

from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
param_grid={'kernel':['linear','poly','rbf','sigmoid'], 'C':[10,100,1000,2000,3000], 'gamma':['scale', 'auto']}
grid=GridSearchCV(SVR(), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
result=grid.cv_results_
```

```python
#print the best parameters after tuning 
print("Best parameters{}:".format(grid.best_params_))
```

```python
#Best Score

print("Best Score: {}".format(grid.best_score_))
```

```python
#Constructing a DataFrame using the dictionary 
table=pd.DataFrame.from_dict(result)
```
