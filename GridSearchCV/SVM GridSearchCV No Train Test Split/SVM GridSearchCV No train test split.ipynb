```python
#importing the necessary library 
import pandas as pd
```

```python
#Read the dataset file
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#Convert the categorical data to NUmerical Data

dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#List the number of columns 
dataset.columns
```

```python
#Input and output split 

independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
independent
```

```python
dependent
```

```python
# Data pre-processing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent=sc.fit_transform(independent)
```

```python
#Model creation 
#Importing GridSearchCV for tuning the hyperparameters 

from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
param_grid={'kernel':['linear','poly','rbf','sigmoid'], 'C':[10,100,1000,2000,3000], 'gamma':['scale', 'auto']}
grid=GridSearchCV(SVR(), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(independent,dependent)
```

```python
#creating a variable named result and storing the results obtained

result=grid.cv_results_
```

```python
#Displaying the best parameters 

print("Best Parameters{}:".format(grid.best_params_))
```

```python
#Constructing a DataFrame using the dictionary 

table=pd.DataFrame.from_dict(result)
```

```python
table
```

```python
#Get the inputs from the user for prediction

Age_input=int(input("Age:"))
BMI_input=float(input("BMI:"))
Children_input=float(input("Children:"))
Sex_male_input=float(input("Male 1 or 0:"))
Smoker_input=float(input("Smoker 1 or 0:"))
```

```python
#Performing Future prediction

Future_prediction=grid.predict([[Age_input,BMI_input,Children_input,Sex_male_input,Smoker_input]])
print("Future_prediction = {}".format(Future_prediction))
```
