```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
#Converting the categorical data into Numerical Data
#Eliminating the dummies from the columns

dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#list the name of the columns

dataset.columns
```

```python
#input split

independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
#output split

dependent=dataset[['charges']]
```

```python
dependent
```

```python
#Model creation
#Importing GridSearchCV for tuning the hyperparameters 

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['squared_error','absolute_error'],
'max_features': ['auto','sqrt','log2'],
'n_estimators':[10,100]}
grid=GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose=3, n_jobs=-1)

## fitting the model for grid search

grid.fit(independent,dependent)
```

```python
#creating a variable named result and storing the results obtained

result=grid.cv_results_
```

```python
#Displaying the best parameters 

print("Best Parameters{}:".format(grid.best_params_))
```

```python
#Best Score

print("Best Score: {}".format(grid.best_score_))
```

```python
#Constructing a DataFrame using the dictionary 

table=pd.DataFrame.from_dict(result)
```

```python
#Get the inputs from the user for prediction

Age_input=int(input("Age:"))
BMI_input=float(input("BMI:"))
Children_input=float(input("Children:"))
Sex_male_input=float(input("Male 1 or 0:"))
Smoker_input=float(input("Smoker 1 or 0:"))
```

```python
#Performing Future prediction

Future_prediction=grid.predict([[Age_input,BMI_input,Children_input,Sex_male_input,Smoker_input]])
print("Future_prediction={}".format(Future_prediction))
```
