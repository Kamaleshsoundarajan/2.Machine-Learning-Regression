```python
#Importing pandas library

import pandas as pd
```

```python
#upload the dataset file

dataset=pd.read_csv("insurance_pre (1).csv")
```

```python
dataset
```

```python
#list the name of the columns

dataset.columns
```

```python
#Mentioning the categorical columns

dataset['sex'] = dataset['sex'].astype ('category')
dataset['smoker'] = dataset['smoker'].astype ('category')
```

```python
dataset
```

```python
#input split

independent=dataset[['age', 'sex', 'bmi','children', 'smoker']]
```

```python
independent
```

```python
#output split

dependent=dataset[['charges']]
```

```python
dependent
```

```python
#train & test set split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model creation

from xgboost import XGBRegressor
regressor=XGBRegressor(enable_categorical=True,n_estimators=300,booster='dart',max_depth=3,learning_rate=0.01)
regressor.fit(x_train,y_train)
```

```python
import xgboost
print(xgboost.__version__)
```

```python
#Evaluating the model

y_pred=regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#saving the model

import pickle
filename="finalized_model_xgboost.sav"
pickle.dump(regressor,open(filename,'wb'))
```
